### Lists

Lists are a **sequence of values** that can be modified by adding, removing, or changing elements. For example, you can append new elements to a list, remove elements from a list, or change the value of an element in a list.

Lists are a mutable data type, which means that they can be changed after being initialized.

Lists are often used to **represent datasets or batches of data**. A dataset is a collection of input-output pairs that are used to train or evaluate a machine learning model.

```python
# Creating a list of numbers
numbers = [1, 2, 3, 4, 5]
print('-'*100)

#NOTE - indexes are weird. The index of the first item in a list
#doesn't start at 1, but 0. Index 1 actually refers to the second item in a list.

# Accessing list elements
print('The first number in the list: ', numbers[0])  # Output: 1
print('The third number in the list: ', numbers[2])  # Output: 3
print('-'*100)

# Slicing a list
print('The second to third numbers in the list: ', numbers[1:3])  # Output: [2, 3]
#output index 1 to 3 (not-included) or second and third item
print('-'*100)

# Modifying list elements
numbers[0] = 10
#single equal sign -> assigning value
#double equal sign -> checking for equality
print('The list with the modified first number: ', numbers)  # Output: [10, 2, 3, 4, 5]
print('-'*100)

# Adding elements to the end of the list
numbers.append(6)
print('The list with a new number added to the rear: ', numbers)  # Output: [10, 2, 3, 4, 5, 6]
print('-'*100)

# Removing elements from a list (not based on idex, but literal item)
numbers.remove(3)
print('The list with number "3" removed from it: ', numbers)  # Output: [10, 2, 4, 5, 6]
print('-'*100)

# Looping through a list
print('Iterating over each number in the list:')
for number in numbers:
    print(number, end=" ")
```

#Numerical Operations

```python
# Integer
a = 5
b = 10

# Float
c = 3.14159
d = 2.71828

# Addition
sum = a + b
print(sum)

# Subtraction
diff = b - a
print(diff)  # Output: 5

# Multiplication
prod = a * b
print(prod)   # Output: 50

# Division
quotient = b / a
print(quotient)  # Output: 2.0 (always returns a float)
print('-'*100)

# Integer division (returns the floor of the division result)
floor_div = b // a
print(floor_div)   # Output: 2

# Modulo (returns the remainder of division)
remainder = b % a
print(remainder)  # Output: 0
print('-'*100)

# Exponentiation
exp = c ** 2
print(exp)  # Output: 9.8695871
print('-'*100)

# Built-in functions
print('Absolute value of -10 is ', abs(-10))  # Output: 10
print('Rounded value of pi is ', round(3.14159, 2))  # Output: 3.14 round to two decimals
print('Maximum values in a list of numbers from 1 to 5 is ', max(1, 2, 3, 4, 5))  # Output: 5
print('Minimum value in a list of numbers from 1 to 5 is ', min(1, 2, 3, 4, 5))  # Output: 1
```

# Loop Statements

Loops are used to execute a block of code **repeatedly**.

There are two main types of loops in Python: for loops and while loops.

Loops are an important part of programming in Python, as they allow you to automate repetitive tasks and **iterate over large datasets**.

It is important to use loops judiciously and optimize them for performance, as they can be computationally heavy, if used improperly.

## `For` loop

`For` loops are used to iterate over a sequence of values, such as a list, tuple, or string.

Below is a code snippet that shows the various examples of loop use.

```python
# Loop over a list
fruits = ["apple", "banana", "cherry"]
print('Printing fruits:')
for fruit in fruits:
    print(fruit)
print('-'*100)

# Loop over a range of numbers
print('Printing numbers:')
for i in range(5): #from 0, 1, 2, 3, 4
    print(i)
print('-'*100)

# Loop over a range of numbers with a step
print('Printing numbers with intervals:') #from 1 to 10, intervals of 2
for i in range(1, 10, 2):
    print(i)
print('-'*100)
```

## `While` loop

`While` loops are used to repeatedly execute a block of code as long as a certain condition is true.

```python
# Count from 1 to 5 using a while loop
print('Count from 1 to 5 using a while loop.')
i = 1
while i <= 5:
    print(i)
    i += 1
print('-'*100)

# Sum the numbers from 1 to 10 using a while loop
print('Sum the numbers from 1 to 10 using a while loop.')
sum = 0
i = 1
while i <= 10:
    sum += i
    i += 1
print("The sum is:", sum)
```

# Conditional Statements

Conditional statements are used in Python to execute different code blocks depending on the value of a certain condition.

## `if-elif-else` statement

The `if-elif-else` statement allows you to test multiple conditions in sequence, and execute different code blocks based on which condition is true.

```python
# Determine the loan eligibility for a given user
income = 60000
credit_score = 725
debt_to_income_ratio = 0.4

if income >= 80000:
    if credit_score >= 750:
        loan_eligibility = "High"
    else:
        loan_eligibility = "Medium"
elif income >= 60000:
    if credit_score >= 700 and debt_to_income_ratio < 0.35:
        loan_eligibility = "High"
    else:
        loan_eligibility = "Medium"
else:
    loan_eligibility = "Low"
```

```python
import io
data = pd.read_csv(io.BytesIO(uploaded['Salary_dataset.csv']))
# Dataset is now stored in a Pandas Dataframe

x = data['YearsExperience']
y = data['Salary']
```

```python
#Data Preprocessing and Cleaning
data['Salary'] = data['Salary']/1000
print(data['Salary'])
data.drop(columns=['Unnamed: 0'], inplace=True)
data.rename(columns={"YearsExperience": 'Experience (years)',
             'Salary': 'Salary (1000 $)'}, inplace=True)
data.head()
```

```python
data.isnull().value_counts()

#Great! No missing values! Usually, we would need to deal with null
#values by either removing those samples completely, filling it in with the average
#or median value of the entire dataset, or the value of its nearby neighbors

```

```python
fig = plt.figure(figsize=(6, 6))

sns.regplot(data=data,
            x='Experience (years)',
            y='Salary (1000 $)',
            color='k', marker='+')
plt.title("Regression Plot: Experience vs. Salary ($1000)")
```

```python
#Model Fitting

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

model = LinearRegression()
x = np.array(data['Salary (1000 $)']).reshape((-1,1))
y = np.array(data['Experience (years)']).reshape((-1, 1))

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

model.fit(x_train, y_train)
print(f'Linear regression R2 value: {model.score(x, y):3.3f}')
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print("R-squared score: ", r2)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error: ", mse)
```

#Random Forest Demo


```python
#Iris-dataset w/ Random Forest

#Import dataset and sklearn model

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import graphviz

#load in dataset

iris = load_iris()
```

```python
#Exploration of dataset

print(iris.target_names)
print(iris.feature_names)
```

```python
#Data visualization - sepal length and width (feature 1 and 2)

#take the first two features (sepal length and width)
X = iris.data[:, :2]
#get all the target y values for each sample
y = iris.target

#get min and max values to plot axis - leave .5 of spare space on each end
#Features
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
#Labels (not the same thing as our y = iris.target, but corresponds to sepal width)
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
```

```python
# we will be creating 2 figures, each of size 8 x 6 dpi
plt.figure(2, figsize=(8, 6))
# clear any previous figures (if any), and make this the current figure
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
#cmap = colormap ; set 1 = built-in color scheme
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
# plt.xticks(())
# plt.yticks(())
plt.title('Red = Setosa, Yellow = Versicolour, Gray = Virginica')
```

##Take a moment to describe the relationships you see in the graph.

```python
# we only take the last two features.
X = iris.data[:, 2:]
# we get all the target for each sample
y = iris.target

# we get minimum and maximum values so that we know what should be plot axis ticks
# (this is only for visualization purpose)
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5

# we get minimum and maximum values so that we know what should be plot axis ticks
# (this is only for visualization purpose)
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

# we will be creating 2 figures, each of size 8 x 6 dpi
plt.figure(2, figsize=(8, 6))
# clear any previous figures (if any), and make this the current figure
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set2)
plt.xlabel("Petal length")
plt.ylabel("Petal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
# plt.xticks(())
# plt.yticks(())
plt.title('Blue = Setosa, Green = Versicolour, Gray = Virginica')

plt.show()
```

##Take a moment to describe the relationships you see in the graph.

```python
#Let's define our train-test split dataset
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2, random_state = 42)
```

##Time to build the actual model, which we can do in like three lines lol.

##You just have to remember **three** keywords
1. Classifier
2. Fit
3. Predict

```python
# Initialize our random forest classifier
forest = RandomForestClassifier(n_estimators = 10) #n_estimators = number of trees
```

```python
# Train our decision tree (tree induction + pruning)
forest = forest.fit(x_train, y_train)
```

```python
#Time to finally predict!

predicted_iris = forest.predict(x_test)
accuracy = accuracy_score(y_test, predicted_iris)
print("Accuracy: " + str(accuracy * 100) + "%")

#Compare results

#predicted
print(predicted_iris)

#actual
print(y_test)

#Ah, they match up! Perfect. Or is it?
#As awesome as 100% accuracy sounds, a perfect score on our metric may actually
#indicate that our model has overfitted (aka does not generalize well to unseen data, our test data)
#But I think our dataset was just really small and thus, the model learned all the noise
#from the data and our testing data was quite similar to the training data
```

```python
#Visualize only one tree (estimator) at once
tree = forest.estimators_[0]
```

```python
dot_data = export_graphviz(tree, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, special_characters=True)

graph = graphviz.Source(dot_data)
graph.render("iris_forest_tree1")
```
